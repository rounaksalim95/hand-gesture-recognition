```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
cap = cv2.VideoCapture(0)
i = 0
while cap.isOpened():
    image, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    #gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    binary_image = cv2.Canny(gray, 100, 200)

    thresh = cv2.threshold(gray, 135, 255, cv2.THRESH_BINARY_INV)[1]
    plt.imshow(thresh, cmap='Greys',  interpolation='nearest')
    plt.show()
    if (cv2.waitKey(1) & 0xFF == ord('q')) or i == 5: break
cap.release()
```
